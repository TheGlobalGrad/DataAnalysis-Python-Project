```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Students-Examscore-Data.csv')
print(df.head())
```

```python
df.isnull().sum()
```

```python
df.columns
```

```python
#Drop unnamed column
df=df.drop("Unnamed: 0", axis=1)
print(df.head())
```

```python
#Change weekly study hours column
df['WklyStudyHours'] = df['WklyStudyHours'].str.replace('5 - 10', '05 - 10')
df.head()
```

# Gender Distribution

```python
sns.set()
plt.figure(figsize=(5,4))
ax = sns.countplot(data=df, x='Gender')
ax.bar_label(ax.containers[0])   #to add data labels on top of bars
plt.xlabel('Gender', fontsize=14, fontweight='bold')
plt.ylabel('Number of Females/Males', fontsize=14, fontweight='bold')
plt.xticks(rotation=0)
plt.show()
```

#### From the above chart I have analysed that the number of Females are more than the number of Males.

```python
gb = df.groupby('ParentEduc').agg({'MathScore':'mean', 'ReadingScore':'mean', 'WritingScore':'mean'})
gb
```

# Impact of Parents' Education

```python
plt.figure(figsize=(4,4))
sns.heatmap(gb, annot = True)
plt.xlabel('Scores', fontsize = 14, fontweight='bold')
plt.xticks(rotation=30)
plt.title('Relationship between Parent\'s Education and Student\'s Score', fontsize=15, fontweight='bold')
plt.show()
```

#### From the above chart I have concluded that Parent's Education has good impact on students' score.

```python
gb1 = df.groupby('ParentMaritalStatus').agg({'MathScore':'mean', 'ReadingScore':'mean', 'WritingScore':'mean'})
gb1
```

```python
plt.figure(figsize=(4,4))
sns.heatmap(gb1, annot = True)  #annot-to show numbers on box
plt.xlabel('Scores', fontsize = 14, fontweight='bold')
plt.xticks(rotation=30)
plt.title('Relationship between Parent\'s Relationship Status and Student\'s Score', fontsize=15, fontweight='bold')
plt.show()
```

#### From the above chart I have concluded that there is no/negligible impact on the students' score due to their parents' marital status.

```python
sns.boxplot(data=df, x='MathScore')  #outliers
plt.title('Outliers for Maths Score', fontsize = 15, fontweight = 'bold')
plt.show()
```

```python
sns.boxplot(data=df, x='ReadingScore')  #outliers
plt.title('Outliers for Reading Score', fontsize = 15, fontweight = 'bold')
plt.show()
```

```python
sns.boxplot(data=df, x='WritingScore')  #outliers
plt.title('Outliers for Writing Score', fontsize = 15, fontweight = 'bold')
plt.show()
```

#### From above charts I have concluded that students are comparatively weak in Maths than Writing and English.

# Distribution of Ethinic Groups

```python
groupA = df.loc[(df['EthnicGroup'] == 'group A')].count()
groupB = df.loc[(df['EthnicGroup'] == 'group B')].count()
groupC = df.loc[(df['EthnicGroup'] == 'group C')].count()
groupD = df.loc[(df['EthnicGroup'] == 'group D')].count()
groupE = df.loc[(df['EthnicGroup'] == 'group E')].count()

l = ['group A', 'group B ', 'group C', 'group D', 'group E']
mlist = [groupA['EthnicGroup'], groupB['EthnicGroup'], groupC['EthnicGroup'], groupD['EthnicGroup'], groupE['EthnicGroup']]
print(mlist)
plt.pie(mlist, labels = l, autopct='%1.2f%%')
plt.title('Distribution of Ethnic groups', fontsize=15, fontweight='bold')
plt.legend()
plt.show()
```

```python
ax = sns.countplot(data = df, x ='EthnicGroup')
ax.bar_label(ax.containers[0])
plt.title('Relationship between Ethnic Group and Student\'s Score', fontsize=15, fontweight='bold')
```

#### The distribution of students by ethnic group is uneven.This chart could be showing how sample size varies by ethnic group  which is important when comparing performance across groups.
